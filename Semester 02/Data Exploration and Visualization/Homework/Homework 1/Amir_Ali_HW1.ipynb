### Task 1: Create a Histogram

```python
plt.hist(dataset['SepalLengthCm'], color = "blue", zorder=1.5)
plt.ylabel('Frequency')
plt.xlabel('Sepal Length')
plt.title('Histogram of Sepal Length(cm)')
plt.grid(zorder=1)
plt.show()
```

```python
plt.hist(dataset['SepalWidthCm'], color = "red", zorder=1.5)
plt.ylabel('Frequency')
plt.xlabel('Sepal Width')
plt.title('Histogram of Sepal Width(cm)')
plt.grid(zorder=1)
plt.show()
```

```python
plt.hist(dataset['PetalLengthCm'], color = "blue", zorder=1.5)
plt.ylabel('Frequency')
plt.xlabel('Petal Length')
plt.title('Histogram of Petal Length(cm)')
plt.grid(zorder=1)
plt.show()
```

```python
plt.hist(dataset['PetalWidthCm'], color = "red", zorder=1.5)
plt.ylabel('Frequency')
plt.xlabel('Petal Width')
plt.title('Histogram of Petal Width(cm)')
plt.grid(zorder=1)
plt.show()
```

### Task 2: Create a Bar Plot

```python
Bar_Plot= dataset.groupby('Species').size().reset_index(name='Frequency')
plt.bar(Bar_Plot['Species'], Bar_Plot['Frequency'], color=['red', 'green', 'blue'])
plt.ylabel('Frequency')
plt.title('Categories of Iris (Class)')
plt.show()
```

### Task 3: Create a Box Plot

```python
# removing Id column 
new_data = dataset[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]] 
print(new_data.head())
plt.figure(figsize = (10, 7)) 
new_data.boxplot()
```

### Task 4: Create a Scatter Plot

```python
scatter_plot_data_grouped = dataset.groupby(['Species'])
colors=['red', 'green', 'blue']

for (name, group), color in zip(scatter_plot_data_grouped, colors):
    x1 = group['SepalLengthCm']
    y1 = group['PetalLengthCm']
    plt.scatter(x1, y1, c=color, zorder=3, label=name)
    z = np.polyfit(x1, y1, 1)
    p = np.poly1d(z)
    plt.plot(x1,p(x1),':', c=color)

plt.grid(zorder=0)
plt.legend()
plt.xlabel('Sepal length')
plt.ylabel('Petal lenght')
plt.title('Sepal lenght(Iris Class) vs Petal lenght(Iris Class)')
plt.show()
```
