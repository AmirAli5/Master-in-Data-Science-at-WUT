## 1.1 Linear Discriminant Analysis (LDA)

### 1.1.1 Implemetation from scratch

```python
# train the model
def train_LDA(X_train,Y_train):
    
    # estimating apriori distributions
    n_0 = (Y_train == 0).sum()
    apriori_0 = n_0 / len(Y_train)
    apriori_1 = 1 - apriori_0
    
    # getting the means
    m_0 = X_train[Y_train == 0].mean(axis = 0)
    m_1 = X_train[Y_train == 1].mean(axis = 0)
    
    # we assume covariance is the same for all classes
    W = np.cov(X_train.T)
    
    return apriori_0, apriori_1, m_0, m_1, W
```

```python
# test the model
def test_LDA(X_test, apriori_0, apriori_1, m_0, m_1, W):
    
    # getting discriminant function
    W_inv = np.linalg.inv(W)
    discr = (m_0-m_1) @ W_inv @ X_test.T - ((m_0-m_1) @ W_inv @ (m_0 + m_1).T)/2 + np.log(apriori_0/apriori_1)
    
    # final prediction
    prediction = [0 if i > 0 else 1 if i < 0 else -1 for i in discr]
    return prediction
```

```python
# Predict the model(LDA)
apriori_0, apriori_1, m_0, m_1, W = train_LDA(X_train,Y_train)
Y_pred = test_LDA(X_test, apriori_0, apriori_1, m_0, m_1, W)
print(Y_pred)
```

```python
# confusion matrix 
from sklearn.metrics import confusion_matrix
confusion_matrix= confusion_matrix(Y_test, Y_pred)
print(confusion_matrix)
#[row, column]
TP = confusion_matrix[1, 1]        
TN = confusion_matrix[0, 0]           
FP = confusion_matrix[0, 1]           
FN = confusion_matrix[1, 0]
```

```python
# accuracy score
accuracy_score1 = (TN + TP)/ (TP + TN + FP + FN)
print(accuracy_score1)
```

## 1.2. Quadratic Discriminant Analysis (QDA)

### 1.2.1 Implemetation from scratch

```python
def train_QDA(X,Y):
    X_mean_0=np.mean(X[Y==0],0)
    X_mean_1=np.mean(X[Y==1],0)
    X_cov_0=np.cov(X[Y==0],rowvar=False)
    X_cov_1=np.cov(X[Y==1],rowvar=False)
    Y_0=1-np.mean(Y)
    Y_1=np.mean(Y)
    X_cov_0_inv=np.linalg.inv(X_cov_0) 
    X_cov_1_inv=np.linalg.inv(X_cov_1) 
    coef_2 = -1/2*(X_cov_1_inv-X_cov_0_inv)
    coef_1 = X_mean_1 @ X_cov_1_inv - X_mean_0 @ X_cov_0_inv
    bias = np.log(Y_1/Y_0) - 1/2*X_mean_1 @ X_cov_1_inv @ X_mean_1.T + 1/2*X_mean_0 @ X_cov_0_inv @ X_mean_0.T + 1/2*np.log(np.linalg.det(X_cov_0)/np.linalg.det(X_cov_1))
    return coef_1, coef_2, bias
```

```python
def test_QDA(X, coef_1, coef_2, bias):
    log_odds=X @ coef_1 + X @ coef_2 @ X.T  + bias
    odds=np.e**log_odds
    prob=1/(1+odds)
    return prob
```

```python
# Predict the model(LDA)
coef_1, coef_2, bias = train_QDA(X_train,Y_train)
Y_pred = test_QDA(X_test, coef_1, coef_2, bias)
print(np.mean(Y_pred > 0.5).astype("int")==Y_test)
```

```python
# confusion matrix 
from sklearn.metrics import confusion_matrix
confusion_matrix= confusion_matrix(Y_test, Y_predict)
print(confusion_matrix)
#[row, column]
TP = confusion_matrix[1, 1]        
TN = confusion_matrix[0, 0]           
FP = confusion_matrix[0, 1]           
FN = confusion_matrix[1, 0]
```

```python
# accuracy score
accuracy_score3 = (TN + TP)/ (TP + TN + FP + FN)
print(accuracy_score3)
```

## 2.1: Schema 1
Training dataset contains n = 1000 observations, p = 2 features and a binary variable that is generated from the Bernoulli distribution with probability of success 0.5. Features of the observations from the class 0 are generated independently from a normal standard distribution. Features of the observations from the class 1 are generated independently from a normal distribution (mean 1, variance 1). LDA assumptions are satisfied

```python
n = 1000
p = 2
y = stats.bernoulli.rvs(p = 0.5, loc = 0, size = n)
X = np.zeros(shape = (n, p))
X[y == 0, :] = np.random.multivariate_normal(np.array([0, 0]), np.array([[1, 0], [0, 1]]), 
                                             size=np.sum(y == 0))
X[y == 1, :] = np.random.multivariate_normal(np.array([1, 1]), np.array([[1, 0], [0, 1]]), 
                                             size=np.sum(y == 1))
```

```python
# lda with scikit-learn
lda = LinearDiscriminantAnalysis()
lda.fit(X, y)
y_pred_lda = lda.predict(X)
```

```python
# qda with scikit-learn
qda = QuadraticDiscriminantAnalysis()
qda.fit(X, y)
y_pred_qda = qda.predict(X)
```

```python
# confusion matrix 
from sklearn.metrics import confusion_matrix
confusion_matrix= confusion_matrix(y, y_pred_qda)
print(confusion_matrix)
#[row, column]
TP = confusion_matrix[1, 1]        
TN = confusion_matrix[0, 0]           
FP = confusion_matrix[0, 1]           
FN = confusion_matrix[1, 0]
```

```python
# accuracy score
accuracy_score_s1 = (TN + TP)/ (TP + TN + FP + FN)
print(accuracy_score_s1)
```

```python
# visualize schema 2 (LDA and QDA)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

## 2.2: Schema 2
Training dataset contains n = 1000 observations, p = 2 features and a binary variable that is generated from the Bernoulli distribution with probability of success 0.5. Features of the observations from the class 0 are generated from a two-dimensional normal distribution (mean 0, variance 1, correlation 0.8). Features of the observations from the class 1 are generated from a two-dimensional normal distribution (mean 1, variance 1, correlation -0.8). LDA assumptions are not satisfied.

```python
n = 1000
p = 2
y = stats.bernoulli.rvs(p = 0.5, loc = 0, size = n)
X = np.zeros(shape = (n, p))
X[y == 0, :] = np.random.multivariate_normal(np.array([0, 0]), np.array([[1, 0.8], [0.8, 1]]), 
                                             size=np.sum(y == 0))
X[y == 1, :] = np.random.multivariate_normal(np.array([1, 1]), np.array([[1, -0.8], [-0.8, 1]]), 
                                             size=np.sum(y == 1))
```

```python
# lda with scikit-learn
lda = LinearDiscriminantAnalysis()
lda.fit(X, y)
y_pred_lda = lda.predict(X)
```

```python
# qda with scikit-learn
qda = QuadraticDiscriminantAnalysis()
qda.fit(X, y)
y_pred_qda = qda.predict(X)
```

```python
# confusion matrix 
from sklearn.metrics import confusion_matrix
confusion_matrix= confusion_matrix(y, y_pred_qda)
print(confusion_matrix)
#[row, column]
TP = confusion_matrix[1, 1]        
TN = confusion_matrix[0, 0]           
FP = confusion_matrix[0, 1]           
FN = confusion_matrix[1, 0]
```

```python
# accuracy score
accuracy_score_s2 = (TN + TP)/ (TP + TN + FP + FN)
print(accuracy_score_s2)
```

## Part 3: Comparison of LDA and QDA methods on simulated data (Repeated ...)

```python
# 1000 observations from Bernouli distribution( p = 0.5)
Y = [stat.bernoulli.rvs(0.5) for i in range(1000)]
```

```python
# 100 times for 20 cor value b/w 0 and 0.99
A = [0.25, 0.5, 0.75, 1]
corr = np.linspace(0, 0.99, 20)
```

```python
QDA = []
LDA = []
for i in range(10):
    for a in A:
        for cor in corr:
            distr0 = stat.multivariate_normal([0, 0], [[1, cor],
                                                       [cor, 1]])
            distr1 = stat.multivariate_normal([a, a], [[1, -cor],
                                                       [-cor, 1]])

            X = [distr0.rvs() if y==0 else distr1.rvs() for y in Y]

            # Fit QDA
            qda = QuadraticDiscriminantAnalysis()
            qda.fit(X, Y)
            
            # Fit LDA
            lda = LinearDiscriminantAnalysis()
            lda.fit(X, Y)
            
            # Predict LDA and QDA and taking mean of all obs
            LDA.append((lda.predict(X) == Y).mean())
            QDA.append((qda.predict(X) == Y).mean())
```

```python
# Accuracy Score
print(f"Accuracy of LDA: {np.mean(LDA)}")
print(f"Accuracy of QDA: {np.mean(QDA)}")
```

```python
plt.figure(figsize = (16, 9))
plt.style.use('fivethirtyeight')

plt.plot(LDA, alpha=0.5, label='lda')
plt.plot(QDA, alpha=0.5, label='qda')

plt.tight_layout()
plt.legend(loc='upper right')
plt.show()
```
