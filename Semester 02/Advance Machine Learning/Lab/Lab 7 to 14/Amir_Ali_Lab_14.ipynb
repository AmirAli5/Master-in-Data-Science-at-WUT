```python
# import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture

plt.style.use('fivethirtyeight')
```

```python
# Generate Dataset
n = 1000
means = [[0,0],[3,3],[4,0]]

def generate_data(s):
    covariance = np.diag([s,s])
    y = np.random.choice([0,1,2], n, p = [0.4, 0.4, 0.2])
    X = np.array([np.random.multivariate_normal(means[i], covariance) for i in y])
    return X, y
```

```python
s = 1
X, y = generate_data(s)
km = KMeans(3).fit(X).labels_
ac = AgglomerativeClustering(3).fit(X).labels_
gm = GaussianMixture(3).fit_predict(X)

plt.scatter(X[:, 0], X[:,1], c = y)
plt.title("True")
plt.show()

plt.scatter(X[:, 0], X[:,1], c = km)
plt.title("Kmeans")
plt.show()

plt.scatter(X[:, 0], X[:,1], c = ac)
plt.title("Hierarchical clustering")
plt.show()

plt.scatter(X[:, 0], X[:,1], c = gm)
plt.title("Gaussian mixture")
plt.show()
```

```python
import re
with open('congress.txt') as f:
    text = f.read()

df = pd.DataFrame([re.split('\s+', line) for line in text.split('\n')])
df.columns = df.iloc[0,:]
df.index = df.iloc[:,0]
df = df.iloc[1:-1, 1:]
```

```python
ac = AgglomerativeClustering(n_clusters=2).fit(df)


from scipy.cluster.hierarchy import dendrogram, linkage
dendrogram(linkage(ac.children_))
```
