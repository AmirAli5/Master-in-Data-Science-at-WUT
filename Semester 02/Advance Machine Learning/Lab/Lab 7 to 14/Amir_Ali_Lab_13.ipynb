### 1. Generate Dataset
Generate 5 datasets using the following schemes

#### Generate Dataset 1

```python
def generate_data_1(n = 1000, p = 50, sigma = 1, low = 0, high = 4, sigma_e = 0.1):
    # sigma - X1~N(0, sigma)
    # low, high - X~U(low, high)
    # sigma_e - e~N(0, sigma_e)
    x1  = np.random.uniform(low, high, n).reshape(n,1)
    x_rest = np.random.normal(0, sigma, n * (p - 1)).reshape(n, p - 1)
    eps = np.random.normal(0, sigma_e, n).reshape(n,1)
    X = np.concatenate((x1, x_rest), axis = 1)
    y = x1 ** 0.5 + eps
    return X, y
```

```python
x1, y1 = generate_data_1()
```

```python
plt.plot(x1[:,0], y1, ".", color = "red")
```

#### Generate Dataset 2

```python
def generate_data_2(n = 1000, p = 50, sigma = 1, low = 0, high = 4, sigma_e = 0.1):
    # sigma - X1~N(0, sigma)
    # low, high - X~U(low, high)
    # sigma_e - e~N(0, sigma_e)
    x1  = np.random.uniform(low, high, n).reshape(n,1)
    x_rest = np.random.normal(0, sigma, n * (p - 1)).reshape(n, p - 1)
    eps = np.random.normal(0, sigma_e, n).reshape(n,1)
    X = np.concatenate((x1, x_rest), axis = 1)
    y = x1 ** 2 + eps
    return X, y
```

```python
x2, y2 = generate_data_2()
```

```python
plt.plot(x2[:,0], y2, ".", color = "red")
```

#### Generate Dataset 3

```python
def hinge(x, cutoff):
    return np.array([i if i > cutoff else 0 for i in x])

def generate_data_3(n = 1000, p = 50, sigma = 1, sigma_e = 0.1):
    # sigma - X1~N(0, sigma)
    # sigma_e - e~N(0, sigma_e)
    X = np.random.normal(0, sigma,size = (n, p))
    eps = np.random.normal(0, sigma_e, size = (n, 1))
    y = hinge(X[:,0], 0).reshape(n,1) + hinge(X[:,0], 1).reshape(n,1) + eps
    return X, y
```

```python
x3, y3 = generate_data_3()
```

```python
plt.plot(x3[:,0], y3, ".", color = "red")
```

#### Generate Dataset 4

```python
def generate_data_4(n = 1000, p = 50, sigma = 1, low = 0, high = 4, sigma_e = 0.1):
    # sigma - X1~N(0, sigma)
    # low, high - X~U(low, high)
    # sigma_e - e~N(0, sigma_e)
    x1  = np.random.uniform(low, high, n).reshape(n,1)
    x_rest = np.random.normal(0, sigma, n * (p - 1)).reshape(n, p - 1)
    eps = np.random.normal(0, sigma_e, n).reshape(n,1)
    X = np.concatenate((x1, x_rest), axis = 1)
    y = np.sin(x1) + eps
    return X, y
```

```python
x4, y4 = generate_data_4()
```

```python
plt.plot(x4[:,0], y4, ".", color = "red")
```

#### Generate Dataset 5

```python
def generate_data_5(n = 1000, p = 50, sigma = 1):
    # sigma - X~N(0, sigma)
    X = np.random.normal(0, sigma, size = (n, p))
    y = X[:,0] < 0
    return X, y
```

```python
x5, y5 = generate_data_5(sigma = 1)
```

```python
plt.plot(x5[:,0], y5, ".", color = "red")
```
