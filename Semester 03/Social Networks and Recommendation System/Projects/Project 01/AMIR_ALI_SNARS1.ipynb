### <font color = #27C3E5> 1.</font> <font color = #41EA46> Square Lattice </font>
Square lattice is a type of lattice in a two-dimensional Euclidean space.

#### <font color = blue> 1.1 Build-in Function.

```python
j = 8
k = 8
sl = networkx.grid_graph((j, k))
networkx.draw(sl)
```

#### <font color = blue> 1.2 Build Own Function.

```python
def square_lattice(a, b):
    sl = networkx.Graph()
    sl.add_nodes_from([(x, y) for x, y in numpy.ndindex(a, b)])
    sl.add_edges_from([((x, y), (x+1, y)) for x, y in numpy.ndindex(a-1, b)])
    sl.add_edges_from([((x, y), (x, y+1)) for x, y in numpy.ndindex(a, b-1)])
    return sl

sl = square_lattice(j, k)
networkx.draw(sl)
```

### <font color = #27C3E5> 2.</font> <font color = #41EA46> Chain </font>

The chain network is a set sequence of messages that are transmitted in a hierarchical order, from the top to the bottom of the chain

#### <font color = blue> 2.1 Build-in Function.

```python
c = networkx.path_graph(j)
networkx.draw(c)
```

#### <font color = blue> 2.2 Build Own Function

```python
def chain(a):
    c = networkx.Graph()
    c.add_nodes_from(range(a))
    c.add_edges_from([(x, x+1) for x in range(a-1)])
    return c

networkx.draw(chain(10))
```

### <font color = #27C3E5> 3.</font> <font color = #41EA46> Complete Graph </font>
In complete graph, every vertex has an edge to all other vertices.

#### <font color = blue> 3.1 Build-in Function.

```python
cg = networkx.complete_graph(j)
networkx.draw_circular(cg)
```

#### <font color = blue> 3.2 Build Own Function

```python
def complete_graph(a):
    cg = networkx.Graph()
    cg.add_nodes_from(range(a))
    cg.add_edges_from(list(itertools.combinations(range(a), 2)))
    return cg

g = complete_graph(10)
networkx.draw(g)
```
